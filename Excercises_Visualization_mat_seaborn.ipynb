#Seaborn Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

#Univariate ---> "one variable" data visualization

Strip Plot

```python
#strip plot
sns.stripplot(x='sex', y='tip', data=tips, jitter=True)
plt.title('Strip plot')
plt.xlabel('Gender')
plt.show()
```

```python
# Grouping with stripplot()
sns.stripplot(x='sex', y='tip', hue='day', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter = False)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tips ($')
plt.show()
```

Swarm Plot

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='day', y='tip', data=tips, facecolor='black')
sns.violinplot(x='day', y='tip', data=tips, color='black')
plt.ylabel('tip ($)')
plt.show()
```

```python
# Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h')
plt.xlabel('tip ($)')
plt.show()
```

Box and Violin Plot

```python
#plt.subplot(1,2,1)
plt.figure(figsize=(10,10))
sns.violinplot(x='day', y='tip', data=tips)
sns.boxplot(x='day', y='tip', data=tips)

plt.ylabel('tips ($)')
plt.show()
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
fig, ax=plt.subplots(nrows=2, ncols=2)
```

```python
plt.hist(tips['total_bill'])
#ax[1].boxplot()
```

```python
#subplots
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tips ($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tips ($)')

plt.show()
```

Combining Plots

```python
sns.violinplot(x='day', y='tip', data=tips, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

#Bivariate ---> "two variables" data visualization

Joint Plot

```python
# Joint Plot

sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

# Joint Plot / Density Plot

```python
# Density plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
# Contours are created representing various levels of density of the data
```

```python
pd.plotting.scatter_matrix(tips)
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("titanic.csv")
```

```python
data.head()
```

#survival - Survival (0 = No; 1 = Yes)
class - Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd) name - Name sex - Sex age - Age sibsp - Number of Siblings/Spouses Aboard parch - Number of Parents/Children Aboard ticket - Ticket Number fare - Passenger Fare cabin - Cabin embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

```python
data.dtypes
```

```python
#Propetion of target (Survived)
data.groupby('Survived')['PassengerId'].count()
```

```python
#Above code can also be written as
data.Survived.value_counts()
```

Visual Exploration

```python
#f, ax = plt.subplots(figsize=(11,5))
plt.figure(figsize=(10,5))
sns.boxplot(x='Survived', y='Age', data=data)
plt.show()
#This shows age doesnt matter among survived
```

```python
#This shows females were given more preference than male
plt.figure(figsize=(10,8))
sns.barplot(x='Sex', y='Survived', data=data);
```

```python
# f, ax = plt.subplots(figsize=(11,5))
plt.figure(figsize=(11,5))
sns.boxplot(x="Sex", y="Age", hue="Survived", data=data);
```

```python
sns.barplot(x='Pclass', y='Survived', data=data)
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=data);
```

```python
sns.barplot(x='SibSp', y='Survived', data=data);
```

```python
sns.barplot(x='Parch', y='Survived', data=data);
```

```python
survived = data.loc[data.Survived==1, "Age"].dropna()
sns.distplot(survived)
plt.title("Survived");
```

```python
not_survived = data.loc[data.Survived==0,"Age"].dropna()
sns.distplot(not_survived)
plt.title("Not Survived")
```

```python
sns.pairplot(data);
```

```python
sns.distplot(a=data[data.Embarked=='C'].Survived,bins=3,kde=False)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()

plt.title("QueensTown")
sns.distplot(a=data[data.Embarked=='Q'].Survived,bins=3,kde=False)
plt.xticks([0,1])
plt.show()

plt.title("Southampton")
sns.distplot(a=data[data.Embarked=='S'].Survived,bins=3,kde=False)
plt.xticks([0,1])
plt.show()

#This shows most of the passengers embarked from Southampton
```
