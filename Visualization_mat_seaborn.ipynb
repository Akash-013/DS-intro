```python
#install and then import matplotlib
# %matplotlib allows graphs to be plotted within the same cell
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = [-3,5,7]
y = [10,2,5]
```

```python
fig = plt.figure(figsize=(8,3))
plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(-3,8)

plt.xlabel('Sales',color='red')
plt.ylabel('product')
plt.title('Line Plot')
plt.suptitle('Sales Comparision', size=20, y=1.05)

#plt.legend
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
plt.plot(x,y)

plt.xlim(0, 10)
plt.ylim(-3, 8)

plt.suptitle('Sales Comparision', size=20, y=1.05)

#plt.legend
plt.subplot(1,2,1)
plt.pie(x)
plt.xlabel('X')
plt.subplot(1,2,2)
plt.pie(y)
plt.xlabel('Y')
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].set_title('Plot 1')
ax[0,0].plot(x,y)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

ax[0,0].plot(x,y,color='green',linestyle='dashed')
ax[0,0].set_title('Line')
ax[1,0].set_title('Scatter Plot')
ax[1,0].scatter(x,y)
ax[1,1].hist(y, color='orange')
ax[1,1].set_title('Histogram')
ax[0,1].set_title('Piechart')
ax[0,1].pie(y)
plt.show()
```

```python
# Practising
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

ax[0,0].plot(x,y,color='#A020F0',linestyle='dotted')
ax[0,0].set_title('Line', size=20, color='#A020F0')
ax[1,0].set_title('Scatter Plot', size=20, color='blue')
ax[1,0].scatter(x,y)
ax[1,1].hist(y, color='yellow')
ax[1,1].set_title('Histogram', size=20, color='#FFC000')
ax[0,1].set_title('Piechart', size=20, color='brown')
ax[0,1].pie(y)
plt.show()
```
